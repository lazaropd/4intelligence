Indique-nos uma amostra da assinatura espectral de um pixel onde a vegetação está saudável, e de um pixel com solo exposto, comparando os valores e como ele se comporta no espectro.

```python
# para trabalhar com imagens espaciais em python, o package rasterio oferece bons recursos
# na Q3(c) será explicada a escolha pelo sentinelhub para download das imagens

import os
import json
import math
import warnings
import rasterio
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
from datetime import datetime, timedelta
from rasterio.plot import show, show_hist

from sentinelhub import SHConfig
from sentinelhub import CRS, BBox, bbox_to_dimensions, MimeType
from sentinelhub import DownloadRequest, SentinelHubDownloadClient

# como as imagens PNG não estão georeferenciadas, o rasterio emite diversos warnings
# para manter uma apresentação mais organizada, omitir os alertas 
# este controle de warnings poderia ser controlado por um parâmetro debug (bool)
warnings.simplefilter('ignore', UserWarning)
```

```python
# análise exploratória: visualização das imagens de referência

aoi_sample_file_rgb = 'inputs/aoi/area-of-interest-aoi-rgb.png'
aoi_sample_file_ndvi = 'inputs/aoi/area-of-interest-aoi-ndvi.png'

sample_rgb = rasterio.open(aoi_sample_file_rgb)
sample_ndvi = rasterio.open(aoi_sample_file_ndvi)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
show(sample_rgb, ax=axs[0,0], adjust=None, title='Area of Interest - RGB')
show(sample_ndvi, ax=axs[0,1], adjust=None, title='Area of Interest - NDVI')
show_hist(sample_rgb.read([1,2,3]), ax=axs[1,0], lw=0.0, stacked=False, 
          label=['R','G','B'], histtype='stepfilled', alpha=0.3, bins=50, title=None)
show_hist(sample_ndvi.read([1,2,3]), ax=axs[1,1], lw=0.0, stacked=False, 
          label=['R','G','B'], histtype='stepfilled', alpha=0.3, bins=50, title=None)
plt.show()
```

## Observações:

- O valor de NDVI originalmente calculado é limitado ao intervalo [-1,1];
    - ele foi transformado para gerar esta imagem em 3 canais;
- Observamos que a imagem PNG do NDVI foi gerada em RGB usando um colormap discreto de RdYlGn;
    - isto pode ser percebido pelas tonalidades de vermelho nas nuvens (NDVI indo para -1);
    - pelas tonalidades de vegetação esparsa/bare soil em amarelos (NDVI próximos de 0);
    - pela vegetação irrigada em tonalidade verde forte (NDVI indo para +1);
    - tanto visualmente no mapa quanto pelo histograma da direita vemos a natureza discreta desta transformação.

```python
# amostrar locais de diferentes classes e comparar suas assinaturas espectrais

classes = {
    'Irrigado': (200, 150),
    'Cultura não irrigada': (60, 150),
    'Solo exposto': (290, 175),
    'Nuvens': (80, 60),
    'Mata nativa/ciliar': (80, 210)
}

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
show(sample_rgb, ax=axs[0], adjust=None, title='Area of Interest - RGB')
show(sample_ndvi, ax=axs[1], adjust=None, title='Area of Interest - NDVI')
for classe, (x, y) in classes.items():
    axs[0].plot([x], [y], marker='o', c='yellow', ms=10, mec='k', mew=3)
    axs[0].text(x, y+10, classe, ha='center', va='top', bbox=dict(fc='yellow', alpha=0.8))
    axs[1].plot([x], [y], marker='o', c='yellow', ms=10, mec='k', mew=3)
    axs[1].text(x, y+10, classe, ha='center', va='top', bbox=dict(fc='yellow', alpha=0.8))
plt.show()
```

```python
# visualização das assinaturas espectrais para os pontos selecionados

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
for classe, (x, y) in classes.items():
    for val in sample_rgb.sample([(x, y)]): 
        axs[0].plot(['Vermelho','Verde','Azul'], val[:-1], lw=5, label=classe)
        axs[0].set_title('Assinatura Espectral - RGB')
    for val in sample_ndvi.sample([(x, y)]): 
        axs[1].plot(['Channel 1','Channel 2','Channel 3'], val[:-1], lw=5, label=classe)
        axs[1].set_title('Assinatura Espectral - NDVI RYG CMAP')
axs[0].set_ylabel('Intensidade de Pixel')
axs[1].legend()
plt.show()
```

## Coleta automatizada das imagens via API

Neste momento, não estamos preocupados ainda com operações em paralelismo, armazenagem em sistemas distribuídos, compactação e outras operações recomendadas para grandes volumes de dados.

Vamos processar as imagens via script automatizado, requests one-at-a-time, usando as APIs abertas disponíveis para o S2 e L8. Existem serviços para contas corporativas e packages com abstrações que permitem coletar e processar imagens em batches conforme será melhor explicado na próxima questão. Bem como a possibilidade de coletar dados brutos na fonte oficial e fazer toda a ETL por conta própria.

Especificamente para esta operação, vamos buscar imagens apenas da nossa AOI usando serviços com alto nível de abstração, e mínimo de recursos de armazenagem (ex.: a coleta dos arquivos .jp2 direto do copernicus demandaria volume de armazenagem muito superiores ao próprio limite de tamanho de arquivo do github para entrega deste desafio).

Em uma aplicação comercial real, seria interessante analisar o tradeoff entre setup inicial rápido, com baixo volume espacial e escalável versus a construção de uma infraestrutura completa de dados.

Uma infraestrutura completa exigiria a coleta, processamento e armazenagem de grandes arquivos (tiles), operações de correção atmosférica, mesclagem de tiles com ajustes de bordas, aplicação de máscaras, recortes, funções de query, armazenamento em cloud e muitas outras. A menos que tal aplicação necessite de uma cobertura completa e ampla de uma grande AOI, pode ser um grande desperdício manter todo um database e utilizar apenas uma pequena fração dele efetivamente.

Abaixo, uma breve comparação de algumas possíveis soluções:

|  | USGS GloVis | ESA Copernicus | Sentinel Hub | Landsat | Google Earth Engine |
|---|---|---|---|---|---|
| Fonte | glovis.usgs.gov | scihub.copernicus.eu | sentinel-hub.com | ers.cr.usgs.gov | developers.google.com |
| Satélites | S2 e L8 | S2 | S2 e L8 | L8 | S2 e L8 |
| Coleta | Manual | API | API | API | API |
| Custo ETL | Complexo | Complexo | Baixo | Complexo | Moderado |
| Custo Storage | Baixo | Alto | Baixo | Alto | Variável |
| Custo SaaS | Nenhum | Nenhum | Escalável | Nenhum | Variável |

**Legenda:**
- ETL: custos com aquisição, processamento e transformação de imagens. Basicamente mão de obra qualificada;
- Storage: custos com armazenamento. Um único tile pode custar até 0,25 US\$/ano em um AWS S3, por exemplo;
- SaaS: sofware como serviço. É possível contratar um serviço altamente abstraído a um custo de 1.000 US\$/ano cobrindo 600 km2/minuto ou 500 requisições/minuto de todas as 13 bandas do S2 usando os serviços do Sentinel Hub, por exemplo. O GEE oferece ainda mais flexibilidade, incluindo VMs, storage, integração com outros produtos google até ferramentas de ML/IA (tensorflow, logit, etc).

## Decisão

Para este desafio, considerando todos os aspectos discutidos acima, escolhi usar a API do Sentinel Hub por oferecer exatamente o que preciso aqui: simultaneamente dados do S2 e L8, abstração, setup simples, sem custo em baixa escala, coleta e armazenagem de imagens de alta resolução da AOI. Para o sentinela, vamos usar a reflectância da superfície (L2A).

Vamos usar um intervalo de 16 dias entre coletas consecutivas no intervalo de 01/01 a 30/06/2020. Neste intervalo, e dadas as frequências de revisitação de ambos, esperamos obter 1 imagem do L8 e 3 imagens do S2 (o algoritmo automaticamente seleciona aquela com menos nuvens neste último caso).

Quando houver mais de 1 imagem disponível no intervalo, será selecionada aquela com menos nuvens. Os resultados serão salvos em um arquivo multispectral, em formato TIFF.

```python
# configurando os parâmetros necessários

load_dotenv()
CLIENT_ID = os.getenv("CLIENT_ID")
CLIENT_SECRET = os.getenv("CLIENT_SECRET")

ENDPOINT = 'https://services.sentinel-hub.com/api/v1/process'
DATASET_PATH = 'dataset/'
RESOLUTION = 10 # maior resolução disponível dentre todas as bandas em coleta

DATE_FROM = datetime(2020, 1, 1)
DATE_TO = datetime(2020, 6, 30)
INTERVAL = 16 # em dias

# conectando ao serviço do Sentinel Hub
config = SHConfig()
if CLIENT_ID and CLIENT_SECRET:
    config.sh_client_id = CLIENT_ID
    config.sh_client_secret = CLIENT_SECRET
if config.sh_client_id == '' or config.sh_client_secret == '':
    print("Por favor informe suas credenciais do Sentinel Hub no arquivo .env")
    print("Em caso de dúvidas, consultar o README do projeto.")
    
print('Sistema de coordenadas do shapefile fornecido:', sorriso.crs)
# converter o polígono para bbox usando o mesmo crs
bbox = BBox(bbox=list(sorriso.geometry.total_bounds), crs=CRS.WGS84)
```

```python
# função reaproveitável e parametrizada para coleta de imagens

def get_request(endpoint, bbox, sensor, dt_from, interval, resolution):
    
    # a API exige que as datas estejam em formato string iso
    dt_to = (dt_from + timedelta(days=interval)).isoformat() + 'Z'
    dt_from = dt_from.isoformat() + 'Z'
    
    # definir o tamanho necessário de imagem para a resolução da imagem de saída
    # cada pixel representa o comprimento em metros dado pelo parâmetro resolution
    # por exemplo, para resolution=10m cada pixel representa 10m reais
    image_size = bbox_to_dimensions(bbox, resolution=resolution)
    
    # anotação das bandas RGB e NIR dos satélites S2 e L8
    # esta API fornece L2A para o S2 mas apenas a imagem não corrigida para o L8
    if sensor == 'S2L2A' or sensor == 'S2L1C':
        bands = ['B02', 'B03', 'B04', 'B08']
        nir = 'sample.B08'
    elif sensor == 'L8L1C':
        bands = ['B02', 'B03', 'B04', 'B05']
        nir = 'sample.B05'
    else:
        bands = []    
    
    eval_script = """
        //VERSION=3
        function setup() {
            return {
                input: %s,
                output: {bands: 4}
            };
        }
        function evaluatePixel(sample) {return [sample.B02, sample.B03, sample.B04, %s];}
    """ % (bands, nir)

    request_dict = {
        "input": {
            "bounds": {
                "properties": {"crs": bbox.crs.opengis_string}, 
                "bbox": list(bbox)
            },
            "data": [{
                    "type": sensor,
                    "dataFilter": {
                        "timeRange": {"from": dt_from, "to": dt_to},
                        "mosaickingOrder": "leastCC"
                    }
            }]
        },
        "output": {
            "width": image_size[0],
            "height": image_size[1],
            "responses": [
                {"identifier": "default", "format": {"type": "image/tiff"}}
            ]
        },
        "evalscript": eval_script
    }
    #print(request_dict)

    # este método automaticamente salva as imagens em GeoTiff para o disco
    # a imagem tiff contém as bandas na seguinte ordem: blue, green, red e nir
    return DownloadRequest(
        request_type = 'POST',
        url = endpoint,
        post_values = request_dict,
        data_type = MimeType.TIFF,
        save_response = True,
        data_folder = DATASET_PATH,
        headers = {'content-type': 'application/json'},
        use_session = True
    )
```

```python
# preparar uma fila de requisições para o serviço de download

s2_list, l8_list = [], []

n_samples = math.ceil((DATE_TO - DATE_FROM).days / INTERVAL)
edges = [DATE_FROM + timedelta(days=(i*INTERVAL)) for i in range(n_samples)]

s2_list = [get_request(ENDPOINT, bbox, 'S2L2A', edge, INTERVAL, RESOLUTION) for edge in edges]
l8_list = [get_request(ENDPOINT, bbox, 'L8L1C', edge, INTERVAL, RESOLUTION) for edge in edges]

s2_list
```

```python
# executar o download usando paralelismo, manter as imagens na memória
# a memória é factível aqui pois as AOIs e o volume de imagens são pequenos

s2_data = SentinelHubDownloadClient(config=config).download(s2_list, max_threads=8)
l8_data = SentinelHubDownloadClient(config=config).download(l8_list, max_threads=8)
```

```python
# visualmente, vemos a evolução das imagens para cada período de tempo

# salvar transformações para próxima etapa da questão
# as imagens RGB serão corrigidas de um fator 3.5 para aumentar a luminosidade das imagens
# o NDVI será calculado como (NIR-RED)/(NIR+RED)
s2_rgb_imgs, s2_ndvi_imgs = [], [] 

# configurar o matplotlib para exibir a sequência de imagens coletadas
ncols = 4
nrows = len(s2_list)
image_size = bbox_to_dimensions(bbox, resolution=RESOLUTION)
aspect_ratio = image_size[0] / image_size[1]

subplot_kw = {'xticks': [], 'yticks': [], 'frame_on': False}
fig, axs = plt.subplots(ncols=ncols, nrows=nrows, 
                        figsize=(3 * ncols * aspect_ratio, 3 * nrows), 
                        subplot_kw=subplot_kw)

# processar e exibir as imagens coletadas
for idx, (s2, l8) in enumerate(zip(s2_data, l8_data)):
    
    with np.errstate(divide='ignore', invalid='ignore'):
        
        # cálculo do NDVI para o S2
        r, nir = s2[:,:,2].astype('float64'), s2[:,:,3].astype('float64')
        s2ndvi = np.where((nir+r)==0., 0, (nir-r)/(nir+r))
        s2_ndvi_imgs.append(s2ndvi)
        # vamos aplicar um fator de correção de 3.5 para aumentar o contraste nas imagens RGB
        s2rgb = np.clip(3.5 * s2[:,:,[2,1,0]], 0, 255).astype('uint8')
        s2_rgb_imgs.append(s2rgb)

        # idem para os dados do L8
        r, nir = l8[:,:,2].astype('float64'), l8[:,:,3].astype('float64')
        l8ndvi = np.where((nir+r)==0., 0, (nir-r)/(nir+r))
        l8rgb = np.clip(3.5 * l8[:,:,[2,1,0]], 0, 255).astype('uint8')
   
    # mostrar mapa    
    if idx == 0:
        axs[idx][0].set_title('Sentinela 2 - RGB')
        axs[idx][1].set_title('Sentinela 2 - NDVI')
        axs[idx][2].set_title('Landsat 8 - RGB')
        axs[idx][3].set_title('Landsat 8 - NDVI')
    axs[idx][0].imshow(s2rgb, vmin=0, vmax=255)
    axs[idx][1].imshow(s2ndvi, cmap='brg', vmin=-1, vmax=1)
    axs[idx][2].imshow(l8rgb, vmin=0, vmax=255)
    axs[idx][3].imshow(l8ndvi, cmap='brg', vmin=-1, vmax=1)
    label = 'de {}\n a {}'.format(datetime.strftime(edges[idx], '%d/%m/%Y'),
                                datetime.strftime(edges[idx]+timedelta(days=INTERVAL), '%d/%m/%Y'))
    axs[idx][0].set_ylabel(label)
```

## Observações:

- enfrentamos bastante restrição atmosférica neste período (especialmente no primeiro trimestre);
- o serviço selecionado não oferece correção de TOA para o L8 automaticamente, imagens esbranquiçadas;
- devido à menor frequência de imagens, a quantidade de imagens úteis (poucas nuvens) do L8 é bem reduzida;
    - vamos simplificar o escopo deste exercício e desprezar as imagens do L8, focando apenas no S2;
    - para responder à Q3(c), vamos selecionar alguns pontos de amostragem;

**Obs.:** os talhões com pivô poderiam ser detectados automaticamente utilizando openCV e detecção de círculos ou métodos de agrupamento de vizinhos próximos como segmentação por Region Growing ou KMeans.  
Outros modelos mais avançados usando segmentação semântica poderiam ser utilizados para detectar e gerar os contornos dos talhões automaticamente, por exemplo usando uma rede neural com topologia U-net. Não vamos entrar neste nível de detalhes aqui para manter o exercício objetivo.

```python
# anotação manual dos pivôs no mapa

sample_rgb, sample_ndvi = s2_rgb_imgs[-1], s2_ndvi_imgs[-1]
# sample_rgb, sample_ndvi = s2_rgb_imgs[1], s2_ndvi_imgs[1]

pivots = {
    'Pivô 01': (150, 500),
    'Pivô 02': (200, 620),
    'Pivô 03': (300, 820),
    'Pivô 04': (250, 300),
    'Pivô 05': (350, 420),
    'Pivô 06': (400, 550),
    'Pivô 07': (450, 700),
    'Pivô 08': (600, 350),
    'Pivô 09': (700, 480),
    'Pivô 10': (650, 650),
    'Pivô 11': (700, 200),
    'Pivô 12': (900, 420),
    'Pivô 13': (950, 500)
}

# para combinar a banda NDVI com bandas RGB, é preciso normalizá-la para a mesma base
# uma combinação de bandas interessante para a agricultura é a color infrared
#   ela combina as bandas ndvi, vermelha e verde
normalized_ndvi = (127.5 + 255*sample_ndvi/2).astype('uint8')
color_infrared = np.dstack((normalized_ndvi, sample_rgb[:,:,0], sample_rgb[:,:,1]))

fig, axs = plt.subplots(1, 3, figsize=(16, 8))
axs[0].imshow(sample_rgb, vmin=0, vmax=255)
axs[1].imshow(sample_ndvi, cmap='brg', vmin=-1, vmax=1)
axs[2].imshow(color_infrared, vmin=0, vmax=255)
axs[0].set_title('AOI - RGB', fontsize=20)
axs[1].set_title('AOI - NDVI', fontsize=20)
axs[2].set_title('AOI - COLOR Infrared', fontsize=20)
for pivot, (x, y) in pivots.items():
    axs[0].plot([x], [y], marker='o', c='yellow', ms=10, mec='k', mew=3)
    axs[0].text(x, y+10, pivot, ha='center', va='top', bbox=dict(fc='yellow', alpha=0.8))
    axs[1].plot([x], [y], marker='o', c='yellow', ms=5, mec='k', mew=2)
    axs[2].plot([x], [y], marker='o', c='yellow', ms=5, mec='k', mew=2)
plt.show()
```

## Seleção das Métricas 

- vamos utilizar os valores de (NIR, RED, GREEN) para enfatizar estas 3 bandas, altamente relacionadas com a agricultura;
    - nesta combinação, a vegetação aparece em vermelhos, solos expostos em marron e nuvens em ciano;
- vamos plotar o valor destas métricas ao longo do tempo para estimar os momentos de plantio e colheita;
    - na questão 2 vimos das assinaturas espectrais que:
        - NDVI positivos indicam vegetação;
        - solos expostos mostram RED >= GREEN;
- vamos utilizar um kernel de tamanho k para reduzir a variabilidade das medições (teorema do limite central);
    - neste caso, ao invés de analisar um único pixel, analisamos um quadrado de tamanho k pixels;
    - por simplifidade, a janela será amostrada para k pixels positivos a partir do ponto amostrado;

```python
# recombinando as bandas de interesse, e organizando em um dataset estruturado

results = []
k = 10

for idx, (rgb, ndvi) in enumerate(zip(s2_rgb_imgs, s2_ndvi_imgs)):
    # normalizar o valor do ndvi para manter mesma dimensão dos dados
    ndvi = (127.5 + 255*ndvi/2).astype('uint8')
    red = rgb[:,:,0]
    green = rgb[:,:,1]
    
    for pivot, (x, y) in pivots.items():
        # o array numpy possui formato (y, x, 3)
        results.append([edges[idx], pivot, ndvi[y:y+k, x:x+k].mean(), red[y:y+k, x:x+k].mean(), green[y:y+k, x:x+k].mean()])

df_results = pd.DataFrame(results)
df_results.columns = ['Data','Pivô','NDVI','RED','GREEN']
    
# criando um filtro rudimentar para separar áreas possivelmente nubladas (cores saturadas)
df_results['Escopo'] = True
df_results.loc[(df_results.RED>200)&(df_results.GREEN>200), 'Escopo'] = False

# remover do escopo áreas com NDVI muito baixos (corpos de água, nuvens densas, etc)
df_results.loc[(df_results.NDVI<50), 'Escopo'] = False

# calcular a relação de verde para vermelho
df_results['GR_RATIO'] = df_results['GREEN'] / df_results['RED']

df_results
```

```python
# é possível visualizar graficamente a evolução estas métricas
# amostrar alguns pivôs menos afetados por nuvens, mantendo a análise mais simples

fig, axs = plt.subplots(1, 1, figsize=(14, 6))

for pivot in pivots.keys():
    subset = df_results.loc[(df_results.Pivô==pivot)&(df_results.Escopo)]
    plt.plot(subset.Data, subset.NDVI, 'o-', lw=3, label=pivot)
plt.legend(loc='lower center', ncol=7)
plt.title('Evolução Temporal - NDVI (transformado -1,1 para 0,255)')
plt.ylim(0, 255)
plt.show()
```
